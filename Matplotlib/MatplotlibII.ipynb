```python
caps=['Hyderabad',
 'Patna',
 'Agartala',
 'Kohima',
 'Chandigarh',
 'Ranchi',
 'Thiruvananthapuram',
 'Mumbai',
 'Bhubaneswar',
 'Chandigarh',
 'Panaji',
 'Gandhinagar',
 'Dispur',
 'Amaravati',
 'Dehradun',
 'Shimla',
 'Bhopal',
 'Jaipur',
 'Aizawl',
 'Srinagar',
 'Kolkata',
 'Raipur',
 'Chennai',
 'Lucknow',
 'Gangtok',
 'Itanagar',
 'Bangalore',
 'Shillong',
 'Imphal']


import requests as w
import json
import pandas as pd
temp=[]
hum=[]
weather=[]
for city in caps:
    try:
    
        data=w.get("http://api.openweathermap.org/data/2.5/weather?q="+city+"&APPID=bafdecd14e244eeadeabf917b9a7b8cf")
    
        data=json.loads(data.content)
        
        p=data['name']+','+str(data['main']['temp'])+','+str(data['main']['humidity'])+','+data['weather'][0]['main']
        print(p,"\n")
        temp.append(data['main']['temp'])
        hum.append(data['main']['humidity'])
        weather.append(data['weather'][0]['main'])
        with open("weather_info.txt",'a') as t:
            t.write(p)
    except:
        print("we are unable to fetch data!")
        
df=pd.DataFrame({"City":caps,"Temperature":temp,"Humidity":hum,"Weather":weather})
print(df)
```

```python
with open("weather_info.csv", 'a') as f:
    f.write(str(df))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
plt.plot(caps,temp,color='red',linestyle='-',linewidth=2,marker='o',markerfacecolor='royalblue',markersize=10)
plt.ylabel('Temperature in Kelvin(K)')
plt.xlabel('State Capitals of India')
plt.xticks(rotation=90)
plt.yticks
plt.title('Temperature variation By State Capitals of India')
```
