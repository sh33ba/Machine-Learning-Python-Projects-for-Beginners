```python
import io
from PIL import Image
import requests as r
from bs4 import BeautifulSoup
data=r.get("https://www.amazon.in/b?ie=UTF8&node=7198569031").content
bs=BeautifulSoup(data,"lxml")
```

```python
i1=bs.find_all("div",{"class":"vxd-wrap"})
for i in range(0,21,2):
    try:
        url=i1[0].find_all("a")[i].find_all("img")[0]["src"]
        d=r.get(url,stream=True).content
        arr2=io.BytesIO(d)
    
        Image.open(arr2).save("Photos/IMage"+str(i)+".jpg")
    except:
        print("*********")
```
