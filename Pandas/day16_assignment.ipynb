```python
u=data1["quality"].unique()
co=[]
m=[]
ma=[]
for i in u:
    y=data1["quality"][data1["quality"]==i].count()
    x=data1["fixed acidity"][data1["quality"]==i].mean()
    z=data1["fixed acidity"][data1["quality"]==i].max()
    co.append(y)
    m.append(x)
    ma.append(z)
```

```python
pd.DataFrame({"quality":u,"count":co,"mean_fixed acidity":m,"max_fixed acidity":ma})
```
