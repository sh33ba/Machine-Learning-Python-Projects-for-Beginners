```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
row,col,i=img.shape
r=[[img[i][j].mean() for j in range(col)] for i in range(row)]
new=np.array(r)
```

```python
plt.imshow(new,cmap='gray')
```
