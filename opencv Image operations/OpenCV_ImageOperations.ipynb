```python
import io
from PIL import Image
import requests as rs
from bs4 import BeautifulSoup
import numpy as np
import glob
import cv2 as cv
```

```python
data=rs.get("https://pixabay.com/images/search/landscape/").content
bs=BeautifulSoup(data,"lxml")
l=len(bs.find_all("div",{"class":"media_list"})[0].find_all("img"))
for i in range(10):
    try:
        url=bs.find_all("div",{"class":"media_list"})[0].find_all("img")[i]["src"]
        im=rs.get(url,stream=True).content
        
        arr=io.BytesIO(im)
        
        Image.open(arr).save("photos/image"+str(i)+".jpg")
                             
                             
    except:
        pass
```
