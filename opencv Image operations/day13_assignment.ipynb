```python
import requests as rs
import io
from PIL import Image
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='https://render.fineartamerica.com/images/rendered/default/poster/8.000/8.000/break/images/artworkimages/medium/1/neon-captain-america-shield-dan-sproul.jpg'
url2='https://static0.srcdn.com/wordpress/wp-content/uploads/2019/05/Avengers-Endgame-Art-Cropped-Captain-America.jpg'
data=rs.get(url,stream=True).content
data2=rs.get(url2,stream=True).content
arr=io.BytesIO(data)
arr2=io.BytesIO(data2)
Image.open(arr).convert('RGB').save("shield.jpg")
Image.open(arr2).convert('RGB').save('cap.jpg')
```

```python
img1=cv.imread('shield.jpg')
img2=cv.imread('cap.jpg')
img1=cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img2=cv.cvtColor(img2,cv.COLOR_BGR2RGB)
```

```python
h,w,d=img2.shape
circle_img=np.zeros((h,w),np.uint8)
cv.circle(circle_img,(int(w/2),int(h/2)),500,1,thickness=-1)
masked_data=cv.bitwise_and(img2,img2,mask=circle_img)
```

```python
plt.imshow(masked_data)
```

```python
import requests as rs
import io
from PIL import Image
import numpy as np
```

```python
url3='https://spaceplace.nasa.gov/nebula/en/nebula3.en.jpg'
url4='https://www.backstageol.com/wp-content/uploads/2019/04/Zoe-620x400.jpg'
data3=rs.get(url3,stream=True).content
data4=rs.get(url4,stream=True).content
arr3=io.BytesIO(data3)
arr4=io.BytesIO(data4)
Image.open(arr3).convert('RGB').save("nebula.jpg")
Image.open(arr4).convert('RGB').save('gamora.jpg')
```

```python
Image.open("nebula.jpg")
```

```python
Image.open('gamora.jpg')
```

```python
img3=cv.imread('nebula.jpg')
img4=cv.imread('gamora.jpg')
```

```python
img3.shape
```

```python
img4.shape
```

```python
crop=img3[91:491,40:660]
crop=cv.cvtColor(crop,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(crop)
```

```python
plt.imshow(img4)
```

```python
b=np.zeros((400,620,3),dtype='uint8')
b1=cv.circle(b,(310,200),200,(255,255,255),-1)
plt.imshow(b1)
```

```python
res1=b1&img4
```

```python
plt.imshow(res1)
```

```python
w=np.ones((400,620,3),dtype='uint8')*255
w1=cv.circle(w,(310,200),200,(0,0,0),-1)
plt.imshow(w1)
```

```python
x=cv.bitwise_xor(res1,w)
```

```python
plt.imshow(x)
```

```python
res2=w1&crop
plt.imshow(res2)
```

```python
y=cv.bitwise_xor(b1,res2)
plt.imshow(y)
```

```python
res3=x&y
plt.imshow(res3)
```
