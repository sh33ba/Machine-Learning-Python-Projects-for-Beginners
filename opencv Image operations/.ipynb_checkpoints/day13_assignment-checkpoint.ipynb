```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import requests as rs
from PIL import Image
import io
```

```python
u='https://nerdist.com/wp-content/uploads/2018/03/Dancing-Groot-Featured.jpg'
data=rs.get(u,stream=True).content
arr=io.BytesIO(data)
Image.open(arr).convert("RGB").save("groot.jpg")
```

```python
img=cv.imread('groot.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
canvas=np.zeros((545,970,3),dtype='uint8')
canvas=cv.circle(canvas,(485,272),272,(255,255,255),-1)
```

```python
plt.imshow(canvas)
```

```python
res=img & canvas
```

```python
plt.imshow(res)
```
